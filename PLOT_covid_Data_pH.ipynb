```python

y1 = covid_actual_df["coviddeaths"]
x1 = covid_actual_df["date"]
plt.plot(x1, y1, label ="Covid Deaths")

y2 = covid_actual_df["average"]
x2 = covid_actual_df["date"]


plt.plot(x2, y2, label = "Traffic Accidents")
#plt.figure(figsize=(x2, y1))

#########   Formats graph
plt.title("Traffic Accidents and Covid Deaths over time In New York Area")
plt.xlabel("Date")
plt.ylabel("Covid Deaths & Traffic Accidents In New York Area")

plt.legend(loc="best")
#ax.legend((y1, y2), ('Covid Deaths', 'Traffic Accidents'),loc='best')

plt.xticks(rotation= 90)

plt.savefig("pictures/Death_Traffic_plot")
```
